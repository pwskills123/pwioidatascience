```python
# create a sample dataset
df = pd.DataFrame({
    'gender': ['male', 'male', 'female', 'female', 'male', 'female', 'male', 'female'],
    'education_level': ['high school', 'college', 'college', 'graduate', 'high school', 'graduate', 'college', 'graduate'],
    'score': [75, 82, 88, 95, 69, 92, 78, 85]
})

# create a crosstab table of gender and education level
ct = pd.crosstab(df['gender'], df['education_level'])
print(ct)
```

```python
# Aggregating Values in Crosstab Table
```

```python
# create a crosstab table of gender and education level with mean score
ct_mean = pd.crosstab(df['gender'], df['education_level'], 
                    values=df['score'], aggfunc='mean')
print(ct_mean)
```

```python
# create a crosstab table of gender and education level with normalization
ct_norm = pd.crosstab(df['gender'], df['education_level'], normalize=True,
                      margins=True)
print(ct_norm)
```

```python
# create a crosstab table of gender and education level with visualization
ct_viz = pd.crosstab(df['gender'], df['education_level'])
ct_viz.plot(kind='bar', stacked=True)
```

Using the iloc[] or loc[] attribute: The iloc[] and loc[] attributes allow you to access rows and columns of a DataFrame using integer positions or labels, respectively. To access the row index of a specific row, use the iloc[] or loc[] attribute and pass the row position or label as the argument.

iloc: Stands for Integer Location

1. It is used for integer-location based indexing.
2. It is primarily integer-based and is used when you want to select data by integer index positions.
3. The syntax for using iloc is df.iloc[row_index, column_index]

loc: Stands for Label Location

1. It is used for label-based indexing.
2. It is primarily label-based and is used when you want to select data by labels (row and column names).
3. The syntax for using loc is df.loc[row_label, column_label].
